```python
# Replace missing values cat - mode, con - mean/median 
def replacer(df):
    cat = list(df.columns[df.dtypes == "object"])

    for col in df.columns:

        if col in cat:
            mode = df[col].mode()[0]
            df[col] = df[col].fillna(mode)
        else:
            mean = df[col].mean()
            df[col] = df[col].fillna(mean)
```

### Apply one hot encoding on X_cat

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(
    handle_unknown="ignore", sparse_output=False
).set_output(transform="pandas")

X_cat_pre = ohe.fit_transform(X_cat)
```

```python
X_cat_pre.head()
```

### Scaling on Numerical (Continuous features)



Standardization is preffered

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().set_output(transform="pandas")

X_con_pre = scaler.fit_transform(X_con)
```

```python
X_con_pre.head()
```

### Combine the categorical and continuous variables

```python
X_pre = X_con_pre.join(X_cat_pre)
X_pre.head()
```

### Step 5 - Train test split

test_size = 0.2 - 0.3

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=21)
```
